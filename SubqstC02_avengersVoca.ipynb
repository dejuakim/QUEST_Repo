```python
# 파일 불러오기
file_path = '/content/sample_data/Avengers.txt' # 파일 경로
with open(file_path, 'r', encoding='cp949') as f:
    avengers = f.read()
```

```python
'''
이 코드는 불러온 텍스트를 전처리 하는 코드입니다.
전처리는 다음의 단계를 가집니다.

1. 문자를 소문자로 변환
2. 모든 특수 기호 제거
3. 단어 단위로 분리
4. 단어의 빈도수 센 뒤, 딕셔너리 형태로 저장
5. 빈도수 순으로 딕셔너리를 내림차순 정렬
6. 정렬 순서대로 인덱스 부여
'''

avengers = avengers.lower()

avengers_txt = ''.join([l for l in avengers if (l.isalnum() or l.isspace())]) # 알파벳/숫자 이거나 공백이면 리스트에 추가

avengers_words = avengers_txt.split()

avengers_dict = {}
for word in avengers_words:
  avengers_dict[word] = avengers_dict.get(word, 1) + 1 # 단어가 있다면 value(빈도) + 1

avengers_sorted = sorted(avengers_dict.items(), key=lambda x: x[1], reverse = True) # 내림차순 정렬

avengers_index = {}
for i in range(len(avengers_sorted)):
  avengers_index[avengers_sorted[i][0]] = avengers_index.get(avengers_sorted[i][0], i) # 정렬 순서대로 인덱스 부여
```

```python
def encoder():
  '''
  사용자가 입력한 텍스트를 input() 으로 받아서
  어벤져스 단어장의 인덱스로 return하는 함수입니다.

  1. 텍스트 입력
  2. 대문자를 소문자로 변환
  3. 띄어쓰기로 단어 구분
  4. 단어장의 단어 인덱스를 확인
  5. 인덱스를 출력
  '''

  txt = input('텍스트를 입력하세요:')
  txt = txt.lower()
  txt_list = txt.split()

  result = []
  for txt in txt_list:
    result.append(avengers_index.get(txt, None)) #단어가 있으면 리스트에 추가
  return result

encoder()
```
