```python
# libararies
import random
```

```python
# 정답숫자 초기화
target_nums = random.sample(range(10),5)

# 5개 값을 입력받기
numbers = []
i = 1
while i < len(target_nums)+1:
  num = int(input(f'''{i}번째 숫자를 입력해주세요(0~9):''')) # 5개 숫자 입력
  if num in numbers: # 중복 숫자라면
    print('중복된 숫자입니다. 다시 입력해주세요')
    continue
  elif num not in range(10): # 0~9 사이의 숫자가 아니라면
    print('0~9 사이의 숫자를 입력해주세요')
    continue
  else: # 올바른 입력이라면
    numbers.append(num) # 숫자 추가
    i += 1

# 값비교
strike = 0
ball = 0
for i in range(len(target_nums)):
  if numbers[i] == target_nums[i]: # 위치&값: strike
    strike += 1
  elif numbers[i] in target_nums: # 위치만: ball
    ball += 1

# 결과 출력
print(f'''strike: {strike}''')
print(f'''ball: {ball}''')
```
