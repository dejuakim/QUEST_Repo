1) 정답 숫자 초기화 - 서로 다른 5개 숫자
2) 사용자에게 숫자 5개 입력받기
3) 정답 숫자와 입력 숫자 비교
- 숫자만 일치 : ball
- 숫자와 위치 모두 일치: strike
4) 결과 출력

```python
from random import *
# 정답 숫자 초기화
ans = []
while len(ans) < 5:
  num = randint(1,9)
  if num not in ans:
    ans.append(num)

while True:

#사용자에게 숫자 5개 입력받기
  num = []
  for i in range(5):
    val = int(input("정수 값을 입력하세요."))
    num.append(val)

# 정답 숫자와 입력 숫자 비교
  strike = 0
  ball = 0
  for i in range(5):
    if ans[i] == num[i]:
     strike +=1
    elif num[i] in ans:
     ball +=1

#결과 출력
  print(f'{strike} strike{ball} ball')

  if strike == 5:
    print("정답입니다!")
    break

```
